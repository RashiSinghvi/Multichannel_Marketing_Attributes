# MACHINE LEARNING MODEL ON DATAFRAME(CHANNEL GROUPING)

```python
CHANNEL=x
```

```python
x
```

```python
CHANNEL=pd.get_dummies(CHANNEL)
```

```python
CHANNEL
```

```python
name=CHANNEL.columns
```

```python
from sklearn import preprocessing
min_max_scaler=preprocessing.MinMaxScaler()
channel_scaled=min_max_scaler.fit_transform(CHANNEL)
channel_scaled=pd.DataFrame(channel_scaled,columns=name)
```

```python
channel_scaled
```

```python
x=channel_scaled.drop(["totalTransactionRevenue"],axis=1)
y=channel_scaled["totalTransactionRevenue"]
```

```python
x_train=x
y_train=y
```

```python
from sklearn.linear_model import LinearRegression as lr
```

```python
linear=lr()
```

```python
linear.fit(x_train,y_train)
```

### ACCURACY PREDICTION FOR TRAIN MODEL


```python
from sklearn.metrics import max_error
pre=linear.predict(x_train)
max_error(pre, y_train)
```

```python

from sklearn.metrics import mean_squared_log_error
mean_squared_log_error(pre, y_train)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train,pre)
```
